```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pylab as plt
```

```python
base  = "https://raw.githubusercontent.com/lmcanavals/"
base += "ml/main/data/"
```

```python
names = ["sepal length", "sepal width",
         "petal length", "petal width", "class"]
data = pd.read_csv(f"{base}iris.data", header=None, names=names)
data.head()
```

```python
data.plot(kind="scatter", x="sepal length", y="petal length")
```

```python
sns.FacetGrid(data, hue="class", height=10) \
.map(plt.scatter, "sepal length", "petal length").add_legend()
```
